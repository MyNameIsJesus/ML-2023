```python
import pandas as pd
import numpy as np
import lxml
```

```python
url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
tables = pd.read_html(url, thousands=".", decimal=",")
t1 = tables[8]
```

```python
##1
t1.head()
```

```python
##2
t1.shape
```

```python
##3
t1 = t1.replace(to_replace="—", value= np.nan)
t1
```

```python
##4
t1.dtypes
```

```python
##5
t1['2014'] = pd.to_numeric(t1['2014'])
t1['2019'] = pd.to_numeric(t1['2019'])
t1.dtypes
```

```python
###6
t1.isnull().sum()
```

```python
##7
t1 = t1.drop(t1.index[len(t1)-1])
t1
```

```python
##8
t2 = t1
for column in t1.columns:
    if(t1.dtypes[column] == "float64"):
      t2[column] = t2[column].fillna(t2[column].mean())
t2
```

```python
##9
t2[t2['2019'] > t2['2019'].mean()]['Регіон']
```

```python
##10

t2[t2['2014'] == t2['2014'].max()]['Регіон']
```

```python
##11
import matplotlib.pyplot as plt
t2['2019'].plot(kind="bar")
```

```python
plt.plot(t2['1950'])
plt.title("Народжуванність за 1950 рік")
plt.show()
```

```python
# t3 = t2.drop(columns=["Регіон"])
plt.pie(t2['1950'], labels=t2["Регіон"])
plt.title('Народжуванність за областями 1950 рік')
plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_trisurf(t2['1950'], t2['2000'], t2['2019'])

ax.set(xticklabels=t2['1950'],
       yticklabels=t2['2000'],
       zticklabels=t2['2019'])
plt.title("Рівень народжуваності")
plt.show()
```

```python
t2['2019'].plot(kind="box", title="Народжуванність в 2019 році тис. людей")
```
