```python
##6
for name in df.columns:
    if(df[name].isnull().sum() != 0 and name != "Язык.программирования"):
        df.pop(name)
df        
```

```python
##8
df = df.dropna()
```

```python
##9
df.shape
```

```python
##10
python_data = pd.DataFrame(data=df[df["Язык.программирования"] == "Python"])
python_data
```

```python
##11
python_data.shape
```

```python
##12
python_data.groupby('Должность')
python_data
```

```python
##13
df1 = pd.DataFrame(data=python_data)
df1.agg({"Зарплата.в.месяц":['max', 'min']})
```

```python
##14
df1['avg'] = ""
def fill_avg_salary():
    df1['avg'] = df1['salary'].mean()
fill_avg_salary()   
df1    
```

```python
##15
df1['avg'].describe()
```

```python
##16
df1.to_csv("dataframe.csv")
```

```python
df1["Изменение.зарплаты.за.12.месяцев"].plot(title="Зміна зарплатні за 12 місяців")
```

```python
import matplotlib.pyplot as plt
df1.groupby('Должность')
plt.pie(df1[df1["Язык.программирования"]=="Python"]['Должность'].value_counts(), labels=df1[df1["Язык.программирования"]=="Python"]['Должность'].unique())
plt.title('Позиції запропонованих вакансій')
plt.show()
```

```python
df1["Зарплата.в.месяц"].value_counts().plot(kind="bar", title="Заробітна плата", figsize=(20,15))
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_trisurf(df1['Возраст'], df1['exp'], df1['salary'])

ax.set(xticklabels=df1['Возраст'],
       yticklabels=df1['exp'],
       zticklabels=df1['salary'])
plt.title("Рівень зарплатні відносно віку та досвіду роботи")
plt.show()
```

```python
df1[df1['Должность']=="Junior Software Engineer"]['salary'].plot(kind="box", title="Зарплатня junior software developer")
```

```python
df1[df1["Язык.программирования"]=="Python"]["Возраст"].value_counts().plot(kind="bar", title="Вік python розробників", xlabel="Вік", ylabel="Кількість людей")
```
