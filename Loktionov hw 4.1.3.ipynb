```python
from matplotlib import pyplot as plt
import random 
import numpy as np
```

```python
colors = ['r','orange','y']
x0=(0,0)
x1=(10,0)
x2=(5, np.sqrt(75))
z = (random.randint(0,10), random.randint(0,10))
```

```python
from IPython import display
%matplotlib inline
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.set_title("Serpinskii figure")
plt.xlabel("X")
plt.ylabel("Y")
plt.ion()
plt.scatter(x0[0], x0[1], c=colors[0], marker="o")
plt.scatter(x1[0], x1[1], c=colors[1], marker="o")
plt.scatter(x2[0], x2[1], c=colors[2], marker="o")
points=[x0,x1,x2]
for i in range(600):
    i = np.random.choice(3)
    z = ((z[0] + points[i][0])/2, (z[1] + points[i][1])/2)
    display.clear_output(wait=True)
    points.append(z)
    colors.append(colors[i])
    for point, color in zip(points,colors):
        plt.scatter(point[0], point[1], c = color, marker="o")
    plt.pause(0.0000000001)
plt.draw()
```
